### Problem 1-1 Load data file and check the data

3. Drop all columns except 'CountryName', 'Date', 'ConfirmedCases', 'ConfirmedDeaths' in the policy dataset and drop the rows containing blank in 'ConfirmedCases' or 'ConfirmedDeaths' columns

```python
###YOUR CODE HERE###
# df4 'policy'
# drop columns except CN, D, CC, CD = JUST EXTRACT four columns
# drop blank-valued rows
df5 = pd.DataFrame(df4)
df5 = df5[['CountryName', 'Date', 'ConfirmedCases', 'ConfirmedDeaths']].dropna()
df5
```

### Problem 1-2 Data Cleaning

1. Fill blank to '' in 'Province_State' column of 'confirmed', 'fatalities', 'metadata' dataset using fillna

```python
###YOUR CODE HERE###
# df1 'confirmed'
# df2 'fatalities'
# df3 'metadata'
filled_df1 = pd.DataFrame(df1).fillna('')
filled_df2 = pd.DataFrame(df2).fillna('')
filled_df3 = pd.DataFrame(df3).fillna('')
```

2. Merge 'Country_Region' and 'Province_State' columns in the confirmed dataset and fatalities dataset, then drop some columns and print the two dataset like the result below 

```python
###YOUR CODE HERE###

# for confirmed dataset
filled_df1['Country_Region'] = filled_df1['Country_Region'] + '_' + filled_df1['Province_State']
filled_df1 = filled_df1.drop(['Province_State', 'Lat', 'Long'], axis = 1)
filled_df1
```

```python
###YOUR CODE HERE###

# for fatalities dataset
filled_df2['Country_Region'] = filled_df2['Country_Region'] + '_' + filled_df2['Province_State']
filled_df2 = filled_df2.drop(['Province_State', 'Lat', 'Long'], axis = 1)
filled_df2
```

3. Count covid19 confirmed cases and fatalities cases for each date by each 'Country_Region' using concat

```python
###YOUR CODE HERE###

# for confirmed dataset

list_CR = []
list_Date = []
list_Confirmed = []
unique_list_CR = filled_df1['Country_Region'].tolist()
unique_list_Date = filled_df1.drop(['Country_Region'], axis = 1).columns.tolist()

for i in range(len(unique_list_CR)):
    list_Date = list_Date + unique_list_Date
    list_Confirmed = list_Confirmed + filled_df1.drop(['Country_Region'], axis = 1).loc[i].values.tolist()
    for j in range(len(unique_list_Date)):
        list_CR.append(unique_list_CR[i])

confirmed_df = pd.DataFrame({'Country_Region':list_CR,
                             'Date':list_Date, 
                             'ConfirmedCases':list_Confirmed})
confirmed_df
```

```python
###YOUR CODE HERE###

# for fatalities dataset

list_CR = []
list_Date = []
list_Fatal = []
unique_list_CR = filled_df2['Country_Region'].tolist()
unique_list_Date = filled_df2.drop(['Country_Region'], axis = 1).columns.tolist()

# create list_CR
for i in range(len(unique_list_CR)):
    list_Date = list_Date + unique_list_Date
    list_Fatal = list_Fatal + filled_df2.drop(['Country_Region'], axis = 1).loc[i].values.tolist()
    for j in range(len(unique_list_Date)):
        list_CR.append(unique_list_CR[i])

fatal_df = pd.DataFrame({'Country_Region':list_CR,
                         'Date':list_Date,
                         'Fatalities':list_Fatal})
fatal_df
```

4. Merge the two dataset in 3. to variable <b>'covid'</b> and split the 'Country_Region' column into 'Country_Region' and 'Province_State' again like the result below

```python
###YOUR CODE HERE###

covid = pd.merge(confirmed_df, fatal_df)
covid.insert(1, 'Province_State', covid['Country_Region'].str.split('_').str[1])
covid['Country_Region'] = covid['Country_Region'].str.split('_').str[0]

covid
```

### Problem 1-4 Data  Grouping
1. Calculate sum of 'ConfirmedCases' and 'Fatalities'  by grouping Countrty_Region
- <b>Hint, Please refer to the following link below<b>
    - __https://pythonspot.com/pandas-groupby/__


```python
###YOUR CODE HERE###
total_covid = covid.groupby(['Country_Region'])[['ConfirmedCases', 'Fatalities']].sum()

total_covid
```

### Problem 1-5 Data Merging
1. Join the table from covid dataset with the metadata table
- __Condition 1 : Check the difference of 'Country_Region' between both datasets__
- __Condition 2 : in the meta dataset due to Province_State, there are many same Country_Region. We use only dataset which not contain province_State data__

```python
###YOUR CODE HERE###

# metadata set with NO Province_State data
```

```python
###YOUR CODE HERE###
id_withPSdata = filled_df3[filled_df3['Province_State'] != ''].index
noPS_df3 = filled_df3.drop(id_withPSdata)
noPS_df3 = noPS_df3.drop(['Province_State'], axis = 1)
join_covid = pd.merge(total_covid, noPS_df3, how='left',on='Country_Region')
join_covid
```

# Problem2 Data Visualizing

## Problem 2-1  plot barchart of  top10 'ConfirmedCases' and 'Fatalities' in joined covid dataset

```python
###YOUR CODE HERE###

# To design the plot, requirements
# X axis for country
# Y axis for amount :: scale: 82613620 -> 8....
top10_join_covid = join_covid.sort_values(by = 'ConfirmedCases', ascending = False)[:10]

xlabels = top10_join_covid['Country_Region'].tolist()
valCF = top10_join_covid['ConfirmedCases'].tolist()
valFatal = top10_join_covid['Fatalities'].tolist()

plt.title('Top 10 Confirmed cases and Fatalities by COVID-19')
plt.bar(np.arange(10), valCF, width=0.4)
plt.bar(np.arange(10) + 0.4, valFatal, width=0.4)
plt.xticks(np.arange(10), xlabels)

plt.show()
```

## Problem2-2  visualizing your data to express 'corona infection situation global map'
 <b> Hint. Refer to the website written at the beginning of the document.<b>

```python
###YOUR CODE HERE###

# global map
# using join_covid in previous problem especially the 'ConfirmedCases'
map = folium.Map()
#max(join_covid['ConfirmedCases']) = 82,613,620
#min(join_covid['ConfirmedCases']) = 1
# we may divide it into 3 layer :: 0 ~ 1,000, 1,000 ~ 1,000,000
for i in range(len(join_covid)):
    latitude = join_covid.loc[i]['lat']
    longitude = join_covid.loc[i]['lon']
    r = join_covid.loc[i]['ConfirmedCases']
    if join_covid.loc[i].isna()['lat'] == False and join_covid.loc[i].isna()['lat'] == False:
        if r >= 10000:
            folium.CircleMarker(location = [latitude, longitude], radius = r/1000000, color='black', opacity = 1, fill=True, weight=1).add_to(map)
        elif r < 10000:
            folium.CircleMarker(location = [latitude, longitude], radius = r/1000, color='gray', opacity = 1, fill=True, weight=1).add_to(map)
map
```

# Problem2-3  visualizing your data to express 'corona infection situation ' according to date

1. Visualize 'ConfirmedCases' trend plot of 'Australia', 'Japan', 'New Zealand', 'Singapore', 'South Korea' from the beginning and the end of policy dataset using iplot

 <b> Hint. Refer to the website written at the beginning of the document.<b>

```python
import cufflinks as cf
cf.go_offline(connected=True)
```

```python
### Visualize each subplot for each country
# using policy dataset :: df5
# trend plot -> using LINE
new_policy = df5.drop(['ConfirmedDeaths'], axis = 1)
df_Aus = new_policy.query("CountryName=='Australia'")
df_Jap = new_policy.query("CountryName=='Japan'")
df_New = new_policy.query("CountryName=='New Zealand'")
df_Sin = new_policy.query("CountryName=='Singapore'")
df_Rok = new_policy.query("CountryName=='South Korea'")

df_Aus.rename(columns={'CountryName':'Aus', 'ConfirmedCases':'Australia'},inplace=True)
df_Jap.rename(columns={'CountryName':'Jap', 'ConfirmedCases':'Japan'},inplace=True)
df_New.rename(columns={'CountryName':'New', 'ConfirmedCases':'New Zealand'},inplace=True)
df_Sin.rename(columns={'CountryName':'Sin', 'ConfirmedCases':'Singapore'},inplace=True)
df_Rok.rename(columns={'CountryName':'Rok', 'ConfirmedCases':'South Korea'},inplace=True)

lp_Aus = df_Aus.iplot(kind = 'line', x='Date', y = 'Australia', title='COVID infection in Australia')
lp_Jap = df_Jap.iplot(kind = 'line', x='Date', y = 'Japan', title='COVID infection in Japan')
lp_New = df_New.iplot(kind = 'line', x='Date', y = 'New Zealand', title='COVID infection in New Zealand')
lp_Sin = df_Sin.iplot(kind = 'line', x='Date', y = 'Singapore', title='COVID infection in Singapore')
lp_Rok = df_Rok.iplot(kind = 'line', x='Date', y = 'South Korea', title='COVID infection in South Korea')
```

```python
### Visualize combined plot of five countries
#new_policy.query("CountryName=='South Korea' or CountryName=='Japan'")

df_combine1 = pd.merge(df_Aus, df_Jap, on = 'Date')
df_combine2 = pd.merge(df_New, df_Sin, on = 'Date')
df_combine3 = pd.merge(df_combine2, df_Rok, on = 'Date')
df_combine = pd.merge(df_combine1, df_combine3, on = 'Date').sort_values(by = 'Date', ascending = False)

lp_Five = df_combine.iplot(kind = 'line', x='Date', y = ['Australia', 'Japan', 'New Zealand', 'Singapore', 'South Korea'], title='COVID infection situation in five countries')
```
